Como se quiere identificar que impacto pueden tener este tipo de variables con las deserciones, se procede con lo siguiente:

```python
# Aplicamos esta función para hacer más fácil las visualizaciones siguientes:

def catplot(varx):
  """ Realiza una gráfica tipo countplot, donde varx es la variable categórica 
  que se desea gráficar en relación con la variable churn. """
  total = float(len(Telecom_dataset[f'{varx}']))
  ax = sns.countplot(x = f'{varx}', hue = "churn", data = Telecom_dataset, palette = 'Paired');
  for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va = 'bottom', size=10)
  plt.show()
```

```python
# Graficamos área (Área geográfica) con relación a churn:
plt.subplots(figsize=(12,7))
sns.countplot(y= "area", hue="churn", data=Telecom_dataset,palette ="Paired");
plt.xticks(rotation = 90)
plt.show()
Telecom_dataset.groupby('area')["churn"].value_counts(normalize=True).unstack(fill_value=0)
```

*A continuación se realiza ese proceso:*

```python
# APLICAMOS EL METODO DE LABEL ENCODING A NUESTRO SET DE DATOS
from sklearn.preprocessing import LabelEncoder

# Agrupamos las variables numéricas y categóricas, cada una en dos nuevos Dataframes
num_data = Telecom_dataset.select_dtypes(include=['int64', 'float64'])
cat_data = Telecom_dataset.select_dtypes(include=['object'])

# Aplicando label encoding sobre las agrupaciones creadas
le = LabelEncoder()
cat_data = cat_data.apply(lambda col: le.fit_transform(col))

final_data = pd.concat([num_data, cat_data], axis=1)
```

```python
final_data.head(5)
```

 **CREANDO NUEVAS FEATURES**

En este paso nos tomamos la tarea de crear nuevas variables a partir de las ya existentes. Entre las nuevas variables estan:


*   **'mean_rev_area' :** se trata del promedio de ingresos por parte de los clientes en cada área donde se registrarón clientes de la empres TELECOM.

*   **'median_rev_area' :** se trata de la mediana de ingresos por parte de los clientes en cada área donde se registrarón clientes de la empres TELECOM.

*  **'median_rev_months' :** Consiste en la mediana de la tarifa de importe promedio por parte de los clientes a partir de el número de meses que ha permanecido el cliente en la compañía.




```python
# Agrupamos los datos de rev_Mean por área donde se encuentran los clientes y calculamos la media de los ingresos de cada área "mean_rev_area"
mean_rev_area = final_data.groupby("area")["rev_Mean"].mean()

# Agrupamos los datos de rev_Mean por área donde se encuentran los clientes y calculamos la mediana de los ingresos de cada área "mean_rev_area"
median_rev_area = final_data.groupby("area")["rev_Mean"].median()

# Agrupamos los datos de rev_mean (Importe de tarifa promedio) por número de meses que ha permanecido el cliente en la compañía
median_rev_months = final_data.groupby("months")["rev_Mean"].median()


# Agregamos nuevas columnas con las agrupaciones anteriores. Los valores son colocados de acuerdo al área que presenta cada cliente
final_data['mean_rev_area'] = final_data['area'].apply(lambda x: mean_rev_area[x] if x > -1 else None)

final_data['median_rev_area'] = final_data['area'].apply(lambda x: median_rev_area[x] if x > -1 else None)

final_data['median_rev_months'] = final_data['months'].apply(lambda x: median_rev_months[x] if x > 1 else None)
```

```python
#Forward selection
def forward_selection(data, target, significance_level=0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while True: #(len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

# **Cross Validation & Hyperparameter Tunning**

Usamos una lista de varios parametros para realizar el cross validation con GridSearch. Luego de ejecutar el modelo, con el comando `.best.estimators_` podemos identificar los mejores parametros para el modelo. De esta manera. escogimos los siguientes parametros para el Modelo LGBMClassifier.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold


n_splits = 6
random_state = 42
lambda_size = 8

# Realizar validación cruzada
cv = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)


# Especificar los parámetros del modelo
params = {
    'objective': ['binary'],
    'metric': ['binary_logloss'],
    'num_leaves': [50, 60, 80],
    'learning_rate': np.logspace(-1, 0.5, 1),
    'feature_fraction': [0.3, 0.5, 0.6, 0.8],
    "n_estimators": [300, 500],
    'bagging_fraction': [0.8],
    'bagging_freq': [5],
    'verbose': [0]
}

modelLGBM_cv = GridSearchCV(
    estimator = lgb.LGBMClassifier(),
    param_grid = params,
    cv=cv,
    verbose = 1,
    scoring="accuracy",
    return_train_score=True,
    n_jobs=-1,
)

modelLGBM_cv.fit(X_train, y_train)
```

```python
from sklearn.inspection import permutation_importance

# Calculamos la importancia de las variables utilizando Permutation importance
result = permutation_importance(modelLGBM_cv, X_train, y_train, n_repeats=10, random_state=0)

# Mostramos los resultados
for i in result.importances_mean.argsort()[::-1]:
    print(f"{X.columns[i]:<8} "
          f"{result.importances_mean[i]:.3f}"
          f" +/- {result.importances_std[i]:.3f}")
    

# Creamos una figura para el plot
fig, ax = plt.subplots()

# Generamos un gráfico de barras para mostrar la importancia de cada variable
ax.bar(X.columns, result.importances_mean)

# Rotamos los nombres de las variables en el eje X para que se lean mejor
plt.xticks(rotation=90)

# Añadimos una etiqueta al eje Y
ax.set_ylabel("Importance")

# Mostramos el plot
plt.show()    
```
